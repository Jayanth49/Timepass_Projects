```python
from sklearn.model_selection import train_test_split

train_set,test_set = train_test_split(data,test_size = 0.2,random_state = 0)

x_train = train_set.drop(["shares"],axis=1)
x_train_labels = train_set["shares"]
x_test = test_set.drop(["shares"],axis=1)
x_test_labels = test_set["shares"]
total_train = data.drop(["shares"],axis=1)
total_train_labels = data["shares"]
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline

num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attributes)),
    ('std_scalar',MinMaxScaler())
])

cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attributes))
])


from sklearn.pipeline import FeatureUnion
full_pipeline = FeatureUnion(transformer_list=[
    ("num_pipeline", num_pipeline),
    ("cat_pipeline", cat_pipeline),
])
```

```python
x_train_prepared = full_pipeline.fit_transform(x_train)
x_test_prepared = full_pipeline.fit_transform(x_test)
total_train_prepared = full_pipeline.fit_transform(total_train)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train_prepared,x_train_labels)
```

```python
from sklearn.metrics import mean_absolute_error

predicted_values = lin_reg.predict(x_train_prepared)
lin_mse = mean_absolute_error(predicted_values,x_train_labels)
lin_mse
```

```python
test_predictions = lin_reg.predict(x_test_prepared)
test_mse = mean_absolute_error(test_predictions,x_test_labels)
test_mse
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor()
regressor.fit(x_train_prepared,x_train_labels)
```

```python
y_predicts = regressor.predict(x_train_prepared)
train_mse =  mean_absolute_error(x_train_labels,y_predicts)
train_mse
```

```python
y_test = regressor.predict(x_test_prepared)
test_mse = mean_absolute_error(x_test_labels,y_test)
test_mse
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [3, 30, 50,100,150], 'max_features': [2, 4, 6, 8]},
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg,param_grid,cv=5,scoring='neg_mean_absolute_error')
grid_search.fit(x_train_prepared,x_train_labels)
```
